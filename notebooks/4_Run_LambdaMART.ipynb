```python
import numpy as np
import pickle
import xgboost as xgb
import graphviz
import matplotlib
import math

path_to_folder = '/Users/rwkoops/PycharmProjects/IR_project/IR_project_02/'
```

```python
training_data = xgb.DMatrix(path_to_folder + 'created_sample/train_set_1m.csv?format=csv&label_column=0') 
testing_data = xgb.DMatrix(path_to_folder + 'created_sample/test_set_1m.csv?format=csv&label_column=0')
```

```python
param = {'max_depth':6, 'eta':0.3, 'objective':'rank:pairwise'}
```

```python
model = xgb.train(param, training_data)
```

```python
xgb.plot_tree(model, num_trees=3)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(150, 100)
```

```python
ax = xgb.plot_importance(model, color='red')
fig = ax.figure
fig.set_size_inches(20, 20)
```

```python
preds = model.predict(testing_data)
```
